```python
from langchain.prompts import PromptTemplate
from langchain_community.llms import Tongyi  # å¯¼å…¥é€šä¹‰åƒé—®Tongyiæ¨¡å‹
import dashscope
import os

# ä»ç¯å¢ƒå˜é‡è·å– dashscope çš„ API Key
api_key = os.environ.get('DASHSCOPE_API_KEY')
dashscope.api_key = api_key
 
# åŠ è½½ Tongyi æ¨¡å‹
llm = Tongyi(model_name="qwen-turbo", dashscope_api_key=api_key)  # ä½¿ç”¨é€šä¹‰åƒé—®qwen-turboæ¨¡å‹

# åˆ›å»ºPrompt Template
prompt = PromptTemplate(
    input_variables=["product"],
    template="What is a good name for a company that makes {product}?",
)

# æ–°æ¨èç”¨æ³•ï¼šå°† prompt å’Œ llm ç»„åˆæˆä¸€ä¸ª"å¯è¿è¡Œåºåˆ—"
chain = prompt | llm

# ä½¿ç”¨ invoke æ–¹æ³•ä¼ å…¥è¾“å…¥
result1 = chain.invoke({"product": "colorful socks"})
print(result1)
```
